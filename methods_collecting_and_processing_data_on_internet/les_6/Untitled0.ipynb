```python
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from keras.datasets import fashion_mnist

from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

(train_imagef,train_labelf),(test_imagef,test_labelf)=fashion_mnist.load_data()
train_imagef = (train_imagef / 127) - 1
test_imagef = (test_imagef / 127) - 1
train_imagef = train_imagef.reshape((-1, 784))
test_imagef = test_imagef.reshape((-1, 784))

model1=Sequential(name='my_model')
model1.add(Dense(128,activation='relu',input_shape=(784,)))
for k in range(2):
  model1.add(Dense(10+(15-k)*10,activation='tanh'))
  #model1.add(Dense(120,activation='tanh'))
model1.add(Dense(10,activation='softmax'))
model1.compile(
  optimizer='adam',
  loss=keras.losses.CategoricalCrossentropy(),
  metrics=['accuracy'],
)

model1.summary()

y_t = to_categorical(train_labelf)
y_tt = to_categorical(test_labelf)

print("Learning model:")
for i in range(2):
  model1.fit(
    train_imagef[:,:],
    y_t[:,:],
    epochs=8,
    batch_size=32, validation_split=0.2
  )
  loss, acc = model1.evaluate(
    test_imagef,
    y_tt
  )

print("Predict:")
y_pred = model1.predict(test_imagef)
y_pred = np.argmax(y_pred, axis=1)

res=0
for i in range(len(y_pred)):
  if y_pred[i]==test_labelf[i]: res+=1
print("Result acc = ",res/len(y_pred))

cm = confusion_matrix(test_labelf, y_pred)
report = classification_report(test_labelf, y_pred, target_names=class_names, digits=4)
print("Classification Report:\n", report)
```
